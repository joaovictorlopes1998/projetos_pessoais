```python
# Importando bibliotecas que vou usar
import os
import glob
import pandas as pd
import chardet
import haversine as hs
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Selecionando o diretório de dados (onde está cada arquivo .csv)
os.chdir("C:\\Users\\joaov\\Desktop\\Torres Consolidado\\Extração de Torres e Escola CSV")
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
```

```python
#Encontrando o enconding correto
#rawdata = open('1.csv', "rb").read()
#chardet.detect(rawdata)
```

```python
#Consolidando todas as bases
combined_csv = pd.concat([pd.read_csv(f, encoding="ISO-8859-1") for f in all_filenames ])
```

```python
#Transformando FreqTxMHz em float, ele vem como object
combined_csv['FreqTxMHz'] = combined_csv['FreqTxMHz'].astype(float)
```

```python
#Dataframe resultante
combined_csv
```

```python
#Criando a função para calcular a diferencia em quilômetros
def calc_km(lat1,lon1,lat2,lon2):
    try:
        loc1=(lat1,lon1)
        loc2=(lat2,lon2)
        return hs.haversine(loc1,loc2)
    except:
        pass
```

```python
#DF de torres filtrando LTE (4G) e o estado
df_torres = combined_csv.loc[(combined_csv['Tecnologia'] == 'LTE') & (combined_csv['FreqTxMHz'] >= 758) & (combined_csv['FreqTxMHz'] <= 894) & (combined_csv['SiglaUf'] == 'PA')]
#DF de escolas filtrando o estado
df_escolas = combined_csv.loc[(combined_csv['Tecnologia'] == 'Escola') & (combined_csv['SiglaUf'] == 'PA')]
```

```python
#Consolidando os dois dataframes para ter 4 colunas (latitude e longitude torre + latitude e longitude escola)
escola_torre_df = df_torres.merge(df_escolas, how='cross')
```

```python
#Criando a coluna de distância entre a torre a escola usando haversine
escola_torre_df['Distancia KM'] = escola_torre_df.apply(lambda x:calc_km(x['Latitude_x'], x['Longitude_x'],x['Latitude_y'], x['Longitude_y']), axis=1)
```

```python
#Dataframe pegando a distância mínima por de escola para torre
escola_torre_df_formatado = escola_torre_df.groupby('NomeEntidade_y').agg({'Distancia KM':[min]}).reset_index()
```

```python
#Criando a flag "Cobertura" onde <= 35km não tem cobertura
escola_torre_df_formatado['Cobertura'] = np.where(escola_torre_df_formatado['Distancia KM'] <= 35, 1, 0)
```

```python
#Média de cobertura
escola_torre_df_formatado['Cobertura'].mean()
```

```python
#Gerando arquivo excel com Escola, quilometragem e Flag Cobertura
escola_torre_df_formatado.to_excel("Cobertura Escolas PA.xlsx")
```
